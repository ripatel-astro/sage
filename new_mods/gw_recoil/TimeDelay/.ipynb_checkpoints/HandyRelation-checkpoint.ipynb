```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import astropy.units as u
```

```python
# DEFINES NUCLEAR STAR CLUSTER SCALING RELATION

def calc_nsc_mass(Mvir):
    if Mvir <= 9 + (.38/.44):
        Mnsc = (.48*(Mvir-9) + 6.51)
    else:
        Mnsc = (.92*(Mvir-9) + 6.13)

    return Mnsc


def calc_t_coal4(Mvir, veldisp, gamma, Rvir):
    Mnsc = calc_nsc_mass(Mvir)
    time = 9.8 - .21*Mnsc + .7*veldisp*np.cos(gamma) - .27*Rvir

    return time

def calc_t_coal5(Mbbh, Mvir, veldisp, gamma, Rvir):
    Mnsc = calc_nsc_mass(Mvir)
    time = 8.2 + .19*Mbbh - .16*Mnsc + .71*veldisp*np.cos(gamma) - .32*Rvir

    return time
```

```python
delay4 = calc_t_coal4(6, 30, 0, 1)
print(delay4)

delay5 = calc_t_coal5(6, 6, 30, 0, 1)
print(delay5)
```

```python
# USING HANDY RELATION EQ 4

num_simulations = 100000

Mvir = np.random.uniform(5, 12, num_simulations) 
veldisp = np.random.uniform(0, 500, num_simulations) 
gamma = np.random.uniform(0, np.pi, num_simulations)
Rvir = np.random.uniform(0, 1, num_simulations)

t_coal = []
for n in range(num_simulations):
    delay = calc_t_coal4(Mvir[n], veldisp[n], gamma[n], Rvir[n])
    t_coal.append(delay)

plt.scatter(Mvir, t_coal, s=0.1)
plt.xlabel(r'log($M_{\star}$)')
plt.ylabel(r'log($T_{coal}$)')
plt.show()

sns.histplot(data=t_coal, binwidth = 20)
plt.xlabel(r'log($T_{coal}$)')
plt.show()
```

```python
# USING HANDY RELATION EQ 5

num_simulations = 100000

Mbbh = np.random.uniform(5, 12, num_simulations)
Mvir = np.random.uniform(5, 12, num_simulations) 
veldisp = np.random.uniform(0, 500, num_simulations) 
gamma = np.random.uniform(0, np.pi, num_simulations)
Rvir = np.random.uniform(0, 1, num_simulations)

t_coal = []
for n in range(num_simulations):
    delay = calc_t_coal5(Mbbh[n], Mvir[n], veldisp[n], gamma[n], Rvir[n])
    t_coal.append(delay)

plt.scatter(Mvir, t_coal, s=0.1)
plt.xlabel(r'log($M_{\star}$)')
plt.ylabel(r'log($T_{coal}$)')
plt.show()

sns.histplot(data=t_coal, binwidth = 20)
plt.xlabel(r'log($T_{coal}$)')
plt.show()
```
