```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import astropy.units as u
```

```python
## DEFINES NUCLEAR STAR CLUSTER SCALING RELATION

#This assumes Mvir is in given in solar masses (adjust formula below if given in other units)
#The division here is the difference between early-stage and late-stage galaxies!

def calc_nsc_mass(Mvir):
    if Mvir <= 9 + (.38/.44):
        Mnsc = (.48*(Mvir-9) + 6.51)
    else:
        Mnsc = (.92*(Mvir-9) + 6.13)

    return Mnsc
```

```python
## TESTS NSC SCALING FORMULA (without scatter)

num_simulations = 100000

Mvir = np.random.uniform(5, 12, num_simulations)

Mstar = []
Mnsc = []
for n in range(num_simulations):
    Mstar.append(Mvir)
    mass = calc_nsc_mass(Mvir[n])
    Mnsc.append(mass)

plt.scatter(Mvir, Mnsc, s=.01)
plt.xlim(6, 12)
plt.xlabel(r'log($M_{\star}$)')
plt.ylabel(r'log($M_{NSC}$)')
plt.axvline(x = 9 + (.38/.44), color='r')
plt.show()

plt.scatter(Mvir, Mnsc-Mvir, s=.01)
plt.xlim(6, 15)
plt.xlabel(r'log($M_{\star}$)')
plt.ylabel(r'log($M_{NSC}/M_{\star}$)')
plt.axvline(x = 9 + (.38/.44), color='r')
plt.show()
    
```

```python
## TESTS NSC SCALING FORMULA (with! scatter)

num_simulations = 10000

Mvir = np.random.uniform(5, 12, num_simulations)
scatter_var = np.random.normal(0, .6, num_simulations)

Mnsc = []
Mnsc_scatter = []
for n in range(num_simulations):
    mass = calc_nsc_mass(Mvir[n])
    Mnsc.append(mass)
    mass_scatter = calc_nsc_mass(Mvir[n]) + scatter_var[n]
    Mnsc_scatter.append(mass_scatter)

plt.scatter(Mvir, Mnsc, s=.01, color = 'cornflowerblue')
plt.scatter(Mvir, Mnsc_scatter, color = 'dodgerblue', s=.1)
plt.xlim(6, 12)
plt.ylim(5, 9)
plt.xlabel(r'log($M_{\star}$)')
plt.ylabel(r'log($M_{NSC}$)')
plt.axvline(x = 9 + (.38/.44), color='r')
plt.show()

plt.scatter(Mvir, Mnsc-Mvir, color='cornflowerblue', s=.01)
plt.scatter(Mvir, Mnsc_scatter-Mvir, color='dodgerblue', s=.1)
plt.xlim(6, 12)
plt.ylim(-4, 0)
plt.xlabel(r'log($M_{\star}$)')
plt.ylabel(r'log($M_{NSC}/M_{\star}$)')
plt.axvline(x = 9 + (.38/.44), color='r')
plt.show()


```

```python
def calc_nsc_mass_small(Mstar):
    Mnsc = (.92*(Mstar-9) + 6.13) 
    return Mnsc
def calc_nsc_mass_large(Mstar):
    Mnsc = (.48*(Mstar-9) + 6.51)
    return Mnsc

n = 100
Mstar_small = np.linspace(5, 11, n)
Mstar_large = np.linspace(7, 11, n)
Mnsc_small = calc_nsc_mass_small(Mstar_small)
Mnsc_large = calc_nsc_mass_large(Mstar_large)

    
plt.plot(Mstar_small, Mnsc_small, color = "black", linestyle = "dashdot")
plt.plot(Mstar_large, Mnsc_large, color = "black", linestyle = "dashdot")
plt.ylim(5, 9)
```
